# **Import Libraries**

```python
import pandas as pd
import numpy as np
from scipy.stats import friedmanchisquare
```

# **File Path & Load**

```python
# File load
model_performance_file = "/content/ADASYN_KFold_Formate_Result(INCART 2-lead Arrhythmia Database).xlsx"
data_reader = pd.ExcelFile(model_performance_file)
all_metrics = data_reader.sheet_names
```

# **Friedman Test**

```python
friedman_results = []

for sheet in all_metrics:
    raw_df = pd.read_excel(data_reader, sheet_name=sheet)

    fold_data = raw_df.iloc[:10]

    target_models = [col for col in fold_data.columns if col.lower() != 'fold']

    model_data_list = [fold_data[model].values for model in target_models]

    try:
        # Friedman Test Calculation
        stat, p_val = friedmanchisquare(*model_data_list)
        friedman_results.append({
            "Evaluation_Metric": sheet,
            "Friedman_Statistic": stat,
            "P_Value": p_val,
            "Significant": p_val < 0.05
        })
    except Exception as e:
        print(f"Error processing {sheet}: {e}")

friedman_df = pd.DataFrame(friedman_results)

print("\n" + "="*75)
print(f"{'FRIEDMAN TEST':^75}")
print("="*75)
print(friedman_df.to_string(index=False))
print("="*75)

output_file = 'Friedman_Test_Results(INCART 2-lead Arrhythmia Database).xlsx'
friedman_df.to_excel(output_file, index=False)
```
